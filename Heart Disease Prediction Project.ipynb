# Split  The Data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('target',axis=1)
y=df['target']
print('Shape of x=',x.shape)
print('Shape of y=',y.shape)
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=75)
print('Shape of X_train = ', x_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', x_test.shape)
print('Shape of y_test = ', y_test.shape)
```

```python
sc = StandardScaler()
sc.fit(x_train)
```

# Using K-Fold

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestRegressor
```

```python
score_lr=cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'), x_train,y_train,cv=3)
print(score_lr)
print("Avg :",np.average(score_lr))
```

```python
score_rf=cross_val_score(RandomForestRegressor(n_estimators=100),x_train,y_train,cv=3)
print(score_rf)
print("Avg:",np.average(score_rf))
```

# K-Fold Cross Validation Techinique

```python
scores1 = cross_val_score(RandomForestRegressor(n_estimators=40),x_train,y_train, cv=10)
print("Avg Score for Estimators=40 and CV=10 :",np.average(scores1))
```

```python
scores2 = cross_val_score(RandomForestRegressor(n_estimators=40),x_train,y_train, cv=10)
print("Avg Score for Estimators=40 and CV=10 :",np.average(scores2))
```

```python
scores2 = cross_val_score(RandomForestRegressor(n_estimators=100),x_train,y_train, cv=10)
print("Avg Score for Estimators=80 and CV=10 :",np.average(scores2))
```
